## Modeling

```python
# Lets us convert into dumpy variables
df=pd.get_dummies(data)
```

```python
# let chech dimension 
print(f"Dumpy dataset rows and columns :",df.shape)
```

```python
# split the dataset
from sklearn.model_selection import train_test_split
```

```python
trainDf,testDf=train_test_split(df,train_size=0.7,random_state=2410)
```

```python
# let us split into dependent and independent variables
trainX=trainDf.drop(["price_range"],axis=1)
trainY=trainDf["price_range"]
testX=testDf.drop(["price_range"],axis=1)
testY=testDf["price_range"]
```

```python
# Modeling 
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier().fit(trainX,trainY)
```

```python
# Prediction
testpredict=rf.predict(testX)
```

```python
# Evaluation
from sklearn.metrics import confusion_matrix,classification_report,r2_score
```

```python
print(confusion_matrix(testY,testpredict))
```

```python
print(classification_report(testY,testpredict))
```

```python
print(f'r2 score for a model which predicts is:',round(r2_score(testY,testpredict)))
```

```python
# Training the model the taining set
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
gnb.fit(trainX,trainY)
```

```python
# Making prediction using datasst
y_predict=gnb.predict(testX)
```

```python
# Modeling 
from sklearn.metrics import accuracy_score
```

```python
accuracy=print(f"Gaussian Naive Banyes Accuracy Score", round(accuracy_score(testY, y_predict)*100))
```

```python
labels = ["low cost", "medium cost", "high cost", "very high cost"]

from sklearn.metrics import ConfusionMatrixDisplay
cm = confusion_matrix(testY, testpredict)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot();
```
